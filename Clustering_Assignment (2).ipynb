# **Theory**

1. **What is unsupervised learning in the context of machine learning?**
- Unsupervised learning is a type of machine learning where the model is trained on unlabeled data—meaning we don’t provide it with predefined outputs or categories. Instead, the algorithm tries to uncover hidden patterns, structures, or groupings in the data all on its own.

2. **How does K-Means clustering algorithm work?**
- K-Means clustering is a popular unsupervised learning algorithm used to group similar data points into K distinct clusters—without needing labeled data.
- Working:
  - Choose the number of clusters (K): You decide how many groups you want the data to be divided into.
  - Initialize K centroids randomly: These are the starting "center points" of each cluster. They can be randomly selected data points or generated within the data range.
  - Assign each data point to the nearest centroid: Using a distance metric (typically Euclidean), each point is assigned to the cluster whose centroid is closest.
  - Update centroids: For each cluster, compute the new centroid by taking the mean of all points assigned to that cluster.
  - Repeat steps 3 and 4: Continue reassigning points and updating centroids until:
  - The centroids stop moving (convergence), or
  - A maximum number of iterations is reached.

3. **Explain the concept of a dendrogram in hierarchical clustering.**
- A dendrogram is a tree-like diagram used to visualize the results of hierarchical clustering, showing how individual data points or clusters are merged step by step based on their similarity.
- Working:
  - In agglomerative clustering (the most common type), each data point starts as its own cluster. The algorithm then:
    - Finds the two closest clusters (based on a distance metric like Euclidean distance).
    - Merges them into a single cluster.
    - Repeats this process until all points are merged into one big cluster.
  - The dendrogram captures this entire process visually.

4. **What is the main difference between K-Means and Hierarchical Clustering?**
- K-Means Clustering
  - Flat clustering: Divides data into K non-overlapping clusters.
  - Requires K upfront: You must specify the number of clusters before training.
  - Iterative refinement: Starts with random centroids and updates them to minimize within-cluster variance.
  - Scalable: Works well with large datasets.
  - Sensitive to initialization: Different runs may yield different results.

- Hierarchical Clustering
  - Hierarchical structure: Builds a tree (dendrogram) of nested clusters.
  - No need to predefine K: You can choose the number of clusters by cutting the dendrogram at a desired level.
  - Deterministic: Produces the same result every time (no randomness).
  - Computationally expensive: Not ideal for very large datasets.
  - Better for exploratory analysis: Helps visualize how clusters form and merge.

5. **What are the advantages of DBSCAN over K-Means?**
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise) offers several advantages over K-Means, especially when dealing with complex or noisy datasets:
  - Unlike K-Means, which requires you to predefine the number of clusters (k), DBSCAN automatically determines the number of clusters based on data density.
  - K-Means assumes clusters are spherical and equally sized. DBSCAN can find clusters of any shape, making it ideal for real-world data that doesn’t fit neat geometric patterns.
  - DBSCAN naturally identifies and excludes outliers as noise points, whereas K-Means tends to assign every point to a cluster—even if it doesn’t belong.
  - DBSCAN groups points based on density connectivity, which allows it to discover clusters in datasets with varying densities—something K-Means struggles with.
  - K-Means can produce different results on different runs due to random initialization. DBSCAN is deterministic, meaning it gives the same result every time (given the same parameters).

6. **When would you use Silhouette Score in clustering?**
- Choosing the optimal number of clusters (K): Run your clustering algorithm (like K-Means) with different values of K and compute the Silhouette Score for each. The K with the highest average score typically gives the best separation.
- Comparing clustering algorithms: If you're trying K-Means, DBSCAN, and Agglomerative Clustering on the same dataset, Silhouette Score helps you compare their effectiveness.
- Validating cluster quality: A high score (close to 1) means points are well matched to their own cluster and far from others. A score near 0 means overlapping clusters. Negative scores suggest misclassified points.

7. **What are the limitations of Hierarchical Clustering?**
- Hierarchical clustering has a high computational cost—typically O(n²) time and space complexity. This makes it impractical for large datasets with thousands or millions of points.
- Outliers can significantly distort the clustering structure, leading to misleading merges or splits. Once a bad merge happens, it cannot be undone, since hierarchical clustering is a greedy algorithm.
- Unlike K-Means, hierarchical clustering doesn’t optimize a specific cost function. This makes it harder to evaluate or compare results quantitatively.
- You must choose a distance metric (e.g., Euclidean, Manhattan) and a linkage method (e.g., single, complete, average), but there’s often no clear guidance on which combination is best. These choices can drastically affect the outcome.
- Hierarchical clustering struggles with datasets that contain both numerical and categorical variables. Calculating meaningful distances across mixed types is non-trivial and often involves arbitrary assumptions.
- Dendrograms are powerful but can be misleading if not interpreted carefully. Small changes in data or linkage method can lead to very different tree structures.

8. **Why is feature scaling important in clustering algorithms like K-Means?**
- Feature scaling is crucial in clustering algorithms like K-Means because the algorithm relies on distance calculations—typically Euclidean distance—to assign points to clusters. If your features are on different scales, the ones with larger ranges will dominate the distance metric, skewing the clustering results.

9. **How does DBSCAN identify noise points?**
- DBSCAN starts by scanning each point’s neighborhood.
- If a point doesn’t have enough nearby neighbors (i.e. it’s not dense enough), and it’s not close to any core point, it’s labeled as noise.
- These points are excluded from all clusters and often represent anomalies or rare cases in the data.

10. **Define inertia in the context of K-Means.**
- In the context of K-Means clustering, inertia is a metric that measures how tightly the data points in a cluster are grouped around their respective centroids. Technically, it's the sum of squared distances between each data point and the centroid of the cluster it belongs to.
- Lower inertia means the clusters are compact and well-separated.
- Higher inertia suggests that points are spread out and the clustering may be less effective.

11. **What is the elbow method in K-Means clustering?**
- The Elbow Method is a visual technique used in K-Means clustering to help determine the optimal number of clusters (K) for your dataset.
- Plot K vs. WCSS. The graph typically shows a sharp drop in WCSS as K increases, followed by a gradual leveling off.
- The point where the drop starts to flatten—forming an “elbow” shape—is considered the optimal K.

12. **Describe the concept of "density" in DBSCAN.**
- In DBSCAN (Density-Based Spatial Clustering of Applications with Noise), "density" refers to how closely packed data points are in a region of the feature space. It’s the core idea behind how DBSCAN identifies clusters and separates noise.
- DBSCAN uses two key parameters to define density:
  - eps (epsilon): The radius of the neighborhood around a point.
  - minPts: The minimum number of points required within that eps-radius to consider the region "dense."

- Types of Points Based on Density
  - Core Point: Has at least minPts neighbors within eps. It lies in a dense region.
  - Border Point: Has fewer than minPts neighbors but is within eps of a core point.
  - Noise Point: Not a core point and not within eps of any core point—too isolated to belong to any cluster.
- Density Connectivity:
  Clusters are formed by chaining together density-connected points:
  - A point is density-reachable from another if there’s a path through core points within eps distance.
  - A cluster is a group of points that are all mutually density-connected.

13. **Can hierarchical clustering be used on categorical data?**
- Yes, hierarchical clustering can be used on categorical data, but it requires some thoughtful adjustments since traditional distance metrics like Euclidean distance don’t apply directly to non-numeric values.

14. **What does a negative Silhouette Score indicate?**
- A negative Silhouette Score indicates that a data point is, on average, closer to points in a different cluster than to points in its own assigned cluster. In simpler terms, it suggests that the point may have been misclassified—it might actually belong to a neighboring cluster rather than the one it was assigned to.

15. **Explain the term "linkage criteria" in hierarchical clustering.**
- In hierarchical clustering, linkage criteria define how the distance between two clusters is calculated when deciding which clusters to merge at each step of the algorithm. It’s the rulebook that guides the clustering process—shaping how the hierarchy (or dendrogram) grows.
- Each method offers a different perspective on what it means for clusters to be “close”:
  - Single Linkage
    - Uses the minimum distance between any two points in the clusters.
    - Tends to form long, chain-like clusters (can suffer from the “chaining effect”).
  - Complete Linkage
    - Uses the maximum distance between any two points.
    - Produces compact, spherical clusters and avoids chaining.
  - Average Linkage
    - Uses the average distance between all pairs of points across the two clusters.
    - Balances the extremes of single and complete linkage.
  - Ward’s Method
    - Merges clusters that result in the smallest increase in total within-cluster variance.
    - Often leads to clusters of similar size and shape.
  - Centroid Linkage
    - Measures the distance between the centroids (means) of the clusters.
    - Can sometimes produce unexpected results due to centroid shifts.

16. **Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?**
- K-Means clustering can struggle with datasets that have clusters of varying sizes or densities because it makes some strong assumptions that don’t always hold true in real-world data. Here's why it can perform poorly in those cases:

  - Assumes Equal Cluster Sizes:
  K-Means tries to minimize the distance between points and their assigned centroids. This works well when clusters are roughly the same size. But if one cluster is much larger, K-Means may:
    - Split the large cluster into multiple smaller ones
    - Merge smaller, distinct clusters into one

  - Assumes Uniform Density:
  K-Means assumes that all clusters have similar density. So if one cluster is dense and another is sparse:
    - The sparse cluster might get ignored or absorbed into a denser one
    - The centroid might be pulled toward the denser region, distorting the boundaries

  - Sensitive to Outliers:
  Outliers can skew centroids, especially in low-density regions. K-Means assigns every point to a cluster—even if it doesn’t belong—so outliers can:
    - Drag centroids away from the true center
    - Create misleading cluster shapes

  - Prefers Spherical Clusters:
  K-Means uses Euclidean distance, which favors spherical, equally sized clusters. It performs poorly on:
    - Elongated or irregularly shaped clusters
    - Overlapping clusters with different spreads

17. **What are the core parameters in DBSCAN, and how do they influence clustering?**
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise) relies on two core parameters that directly shape how clusters are formed:

  - eps (Epsilon):
    This defines the radius of the neighborhood around a data point. In other words, it sets how close points need to be to be considered neighbors.
    - Smaller eps → tighter clusters, more points labeled as noise.
    - Larger eps → looser clusters, risk of merging distinct groups.
    - Tip: Use a k-distance plot to find a good eps—look for the “elbow” in the curve.

  - minPts (Minimum Points):
    This is the minimum number of points required within the eps radius to form a dense region (i.e., a cluster).
    - Low minPts → more clusters, possibly including noise.
    - High minPts → fewer, denser clusters; may miss smaller ones.
    - Rule of thumb: Set minPts ≥ D + 1, where D is the number of dimensions in your data.


18. **How does K-Means++ improve upon standard K-Means initialization?**
- K-Means++ improves upon standard K-Means by choosing smarter initial centroids, which leads to better clustering results and faster convergence. Here's how it makes a difference:

  - How K-Means++ Fixes This
    K-Means++ selects initial centroids using a probabilistic, distance-aware strategy:
    - Pick the first centroid randomly from the data points.
    - For each remaining point, compute its distance to the nearest chosen centroid.
    - Select the next centroid with probability proportional to the square of that distance.
    - Repeat until K centroids are chosen.
    This ensures that new centroids are spread out, reducing overlap and improving separation between clusters.

19. **What is agglomerative clustering?**
- Agglomerative clustering is a bottom-up hierarchical clustering method where each data point starts in its own cluster, and pairs of clusters are merged step by step based on their similarity until all points belong to a single cluster or a stopping condition is met.
- Use Cases
  - Customer segmentation
  - Document or gene clustering
  - Exploratory data analysis

20. **What makes Silhouette Score a better metric than just inertia for model evaluation?**
- While both inertia and Silhouette Score are used to evaluate clustering quality—especially in algorithms like K-Means—they offer very different insights. Here's why the Silhouette Score is often considered more informative:

  - Inertia: Measures Compactness Only
    - Definition: Sum of squared distances between each point and its assigned cluster centroid.
    - Limitation: It always decreases as you increase the number of clusters, which can be misleading.
    - Use Case: Best used with the Elbow Method to find a point of diminishing returns.

  - Silhouette Score: Balances Cohesion and Separation
    - Definition: Measures how similar a point is to its own cluster (cohesion) versus other clusters (separation).
    - Range: From -1 to +1
    - +1: Well-clustered
    - 0: On the boundary
    - -1: Possibly misclassified
    - Advantage: Gives a more holistic view of clustering quality by considering both intra-cluster tightness and inter-cluster separation.

29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
# Synthetic Data
from sklearn.datasets import make_blobs
x,y = make_blobs(n_samples=250, centers=4, cluster_std=1.0, random_state=1)

# Standardize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# clustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
k = range(2,6)
silhouette_scores = []
for i in k:
  model = KMeans(n_clusters=i, random_state=1)
  labels = model.fit_predict(x_scaled)
  score = silhouette_score(x_scaled, labels)
  silhouette_scores.append(score)

# Visuals
import matplotlib.pyplot as plt
plt.bar(k,silhouette_scores, color = 'purple')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering (k = 2 to 5)')
plt.xticks(k)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
# Synthetic Data
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=250, centers=4, cluster_std=2.0, random_state=1) # high std >> more overlap

# K Means
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)
centers = kmeans.cluster_centers_

# Visualization
import matplotlib.pyplot as plt
import numpy as np

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='Pastel2')
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', edgecolor='k')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("K-Means Clustering with Overlapping Blobs and Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

34.  Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = load_breast_cancer()
X = data.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Compute inertia for K = 2 to 6
inertias = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Visualization
plt.figure(figsize=(8, 5))
plt.plot(k_values, inertias, marker='o', linestyle='-', color='teal')
plt.title("K-Means Inertia for K = 2 to 6 (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(k_values)
plt.grid(True)
plt.tight_layout()
plt.show()
```

43. Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate noisy circular data
X, y = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Visualize side-by-side
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# KMeans plot
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='coolwarm', s=50, edgecolor='k')
axes[0].set_title("KMeans Clustering")

# DBSCAN plot
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='coolwarm', s=50, edgecolor='k')
axes[1].set_title("DBSCAN Clustering")

for ax in axes:
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

44.  Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute silhouette scores for each sample
silhouette_vals = silhouette_samples(X, labels)

# Plot the silhouette scores
plt.figure(figsize=(8, 6))
y_lower = 10
for i in range(3):
    ith_cluster_silhouette_vals = silhouette_vals[labels == i]
    ith_cluster_silhouette_vals.sort()
    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_vals,
                      alpha=0.7, label=f'Cluster {i}')
    y_lower = y_upper + 10

plt.axvline(x=silhouette_score(X, labels), color="red", linestyle="--", label="Average Silhouette Score")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Sample Index")
plt.title("Silhouette Plot for KMeans Clustering (Iris Dataset)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

48.  Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dimensions using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agglo.fit_predict(X_tsne)

# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50, edgecolor='k')
plt.title("Agglomerative Clustering on Digits Dataset (t-SNE Reduced)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
